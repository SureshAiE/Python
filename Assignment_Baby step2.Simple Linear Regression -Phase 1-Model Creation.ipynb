```python
#AI in HR Department : To Predict the salary of 13 years of experienced employee.

#By Analysing the provided Data Set from Salary_Data.csv
#Here we have input and output data is very clear and data has Numbers and it is also continous one
#Hence Stage1:Domain=Machine Learning  Stage2:Learning=Supervised Learning   Stage3:Category=Regression

#Steps for Model Creation/Learning-Phase 1:
#1.Data Collection, 2.Data Preprocessing, 3.Input/Output Split, 4.Split Train and Test, 5.Model Creation by using Train Set Data, 
#6.After creating Model we will check with Test Set Data, 7.Then we will Evaluate Metrics to find out the Closeness of Accuracy and its performance
#7.After Knowing the Successful performance of this model, can be saved as Model in Pickle
```

```python
#Pandas is a Python package that helps you manipulate and analyze data, especially tabular data in the form of rows and columns.
#It's a popular tool for data wrangling, which is the process of transforming data from unusable to usable formats for analytics.
import pandas as pd
```

```python
#A dataset in Python is a collection of data that is organized and structured for analysis, visualization, and manipulation. 
#Datasets are a fundamental tool in data analytics, machine learning, and data analysis. 
#They can contain many types of data, including numerical values, text, images, and audio recordings.
#here pd is a class and read_csv is a function and Salary_Data.csv is a dataset

dataset=pd.read_csv("Salary_Data.csv")
```

```python
#Now we can see what data are in dataset by calling this variable
dataset
```

```python
#Now we can take that single input from dataset table
independent=dataset[["YearsExperience"]]
```

```python
##Now we can see what data are in independent by calling this variable
independent
```

```python
#Now we can take that single Output from dataset table
dependent=dataset[["Salary"]]
```

```python
#Now we can see what data are in dependent by calling this variable
dependent
```

```python
#Now we have done input and output split from dataset as per Model Creation/Learning step
#Then we have to split train data and test data from dataset in phase 1

#Now to do this task we have to access "sklearn" class with "model_selection" function's File folder
#Then we particularly import train_test_split function

from sklearn.model_selection import train_test_split

#Now we have to pass the parameters for train_test_split functions
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Here we have to seperate input and out train and test variables to access the sklearn class
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Now we have to create Model and learning. sklearn is an algorithm library. from this we will call LinearRegression function
from sklearn.linear_model import LinearRegression
```

```python
#After calling the class we will assign variable as regressor
regressor=LinearRegression()
```

```python
#Now we will call fit function from regressor to know the value of w-weight(slope-intervel-gap) and b-bias(intercept-Origin-Minimum value)
regressor.fit(X_train,y_train)
```

```python
#Now we can call weight function to know the value of w
weight=regressor.coef_
```

```python
weight
```

```python
#Now we can call bias function to know the value of b
bias=regressor.intercept_
```

```python
bias
```

```python
#Now we have created model and got the values of w and b (y=wX+b), then we will have to Evaluate (Cross Checking) whether it is error free or more error.
#close to the value of 1 = Error free
#close to the value of 0 = More Error
y_pred=regressor.predict(X_test)
```

```python
#Applying Evaluation Formulae
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#Model is working well as the score is close to the value of 1
```

```python
#Now we have to save the model to send for final use
import pickle
filename="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
#Now saved model can be loaded for checking
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
