```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['Age', 'Bmi', 'Children', 'Sex_Male', 'Smoker_Yes']]
independent
```

```python
dependent=dataset[['Charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators = 50,criterion='absolute_error',max_features='sqrt',random_state=0)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle 
filename="finalized_model_RandomForest1.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_RandomForest1.sav",'rb'))
result=loaded_model.predict([[58,28,3,False,True]])
```
